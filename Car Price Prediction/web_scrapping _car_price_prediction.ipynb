```python
# Importing Libraries/Dependencies
import pandas as pd
import numpy as np
import time
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.chrome.service import Service
import requests
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
# Connecting to the webdriver
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://www.cardekho.com/used-cars+in+ahmedabad"
driver.get(url)
time.sleep(6)
```

```python
brand=[]
model=[]
variant=[]
manuf_year=[]
Driven_KM=[]
fuel=[]
location=[]
No_owners=[]
price=[]
```

```python
time.sleep(3)

start=0
end=4
for page in range(start,end):
    
    brands=driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    for i in brands:
        brand.append(i.text)
        
    prices=driver.find_elements_by_xpath("//span[@class='amnt ']")
    for i in prices:
        price.append(i.text)
        
    models=driver.find_elements_by_xpath("//div[@class='car_variant']")
    for i in models:
        model.append(i.text)  
        
    kms=driver.find_elements_by_xpath(("//div[@class='truncate dotlist-2']/div[1]"))
    for i in kms:
        Driven_KM.append(i.text)
        
    fuels=driver.find_elements_by_xpath(("//div[@class='truncate dotlist-2']/div[2]"))
    for i in fuels:
        fuel.append(i.text)
        
    variants=driver.find_elements_by_xpath(("//div[@class='truncate dotlist-2']/div[3]"))
    for i in variants:
        variant.append(i.text)
```

```python
df1=pd.DataFrame({
'Brand':brand[:],                 
'Price':price[:],
'Model':model[:],
'KMS_driven':Driven_KM[:],
'Fuel':fuel[:],
'Variant':variant[:]})
df1
```

```python
driver.close()
```
